```python
## this function take text and qustion and return answer  
# based on our model to prediction 
# we need 1-input_ids 
#         2-attention_mask 
# as input to our model  and we will get it from Tokenizer 
## then we need to specifice start char of our answer and it end ---> it given from model as output for it 
## let do it 



### creat function  
import numpy as np
def answer_by_longformer(context , question):
  tokens = tokenizer.encode_plus(context , question , return_tensors ='pt') 
  input_ids = tokens['input_ids'] 
  attention_mask = tokens['attention_mask'] 
  start_token , end_token = model(input_ids , attention_mask=attention_mask )
  # now we have start of answer token and it end 
  # now we need to specifice it in our main context 
  # and then convert it by decoder 
  # finally return our answer as output of this fucntion 

  all_tokens = tokenizer.convert_ids_to_tokens(input_ids[0].tolist()) 
  # specifice our tokens answer :
  answer_tokens = all_tokens[torch.argmax(start_token) : torch.argmax(end_token)+1]
  final_answer = tokenizer.decode(tokenizer.convert_tokens_to_ids(answer_tokens))
  return final_answer 
```
